```python
import numpy as np

#1차원 벡터
a=np.array([1,2,3,4,5,6]) 
print(a)

#2차원 매트릭스
A=np.array([[1,2,3],[4,5,6]])
print(A)

#array 하나의 형식 통일 
B=np.array([[3,4,5],[6,7,8.0]])
print(B) #실수로 통일

print(A+B) #array 됨

print(A.ndim) #몇 차원 2
print(A.shape) #행,열 개수 (2,3)
print(A.size) #원소 개수 6

print(A.T) #transpose
print(A.dtype) #int32
print(B.dtype) #float64
```

**p7 개념**

- reshape
1. 원소 갯수/행 갯수=n
2. 벡터에서 원소 n개씩 묶음

```python
A = np.arange(1, 21)
print(A)
B = A.reshape(4,5) #5개씩 묶음
print(B)
C=B.reshape(2,10) #10개씩 묶음
print(C)

#C=D 같음
#D=A.reshape(2,10)
#print(D)
#C==D
```

**p8 slicing**

```python
#1차원 벡터
a=[1,3,5,7,9]

#slicing은 차원 유지
a[:3]
a[1:3] 
a[-2:] #뒤 2번째~끝
a[2:3] 

#indexing 차원 감소
a[2] #원소 반환
```

```python
#2차원 매트릭스
print(B) #한 행 = 한 원소 인식

#slicing 차원 유지
print(B[:2])
print(B[-2:])
print(B[1:2])

#indexing 차원 감소
print(B[1])
```

```python
B[:2]==B[:2,:]
B[1:2]==B[1:2,:]
```

```python
print(B)
#차원 유지
print(B[:, -2:])
print(B[:, :2])
print(B[:, 1:2])

#차원 축소
print(B[:, 1])
```

```python
#p10
a=np.arange(0, 20)
a=a.reshape(5,4) #a 저장
print(a)
```

```python
print(a[[0, 1, 3],:])
print(a[[0, 1, 3], 2]) #열 하나 추출 ㄱㅊ
```

```python
print(a[[0,1,3], [0,2]]) #에러, 열 여러개 추출 ㅂㄱㄴ
a[[0,1,3], :][:, [0,2]] #행 추출 -> 열 추출
```

```python
A = np.arange(1,21)
B = A.reshape([4,5])
print(B)
print(B[[0,1], :])
print(B[:, [3,4]])
```
